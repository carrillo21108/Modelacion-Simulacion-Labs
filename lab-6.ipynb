### Ejercicio #1

```julia
using Pkg
Pkg.add("JuMP")
Pkg.add("GLPK")
```

```julia
# Importing JuMP and GLPK
using JuMP
using GLPK

# Create the model with the GLPK solver
model = Model(GLPK.Optimizer)
```

```julia
# Decision variables
@variable(model, x1 >= 0)  # chamarras
@variable(model, x2 >= 0)  # relleno de plumas
@variable(model, x3 >= 0)  # pantalones
@variable(model, x4 >= 0)  # guantes

# Objective function: maximize profit (adjusted coefficients and constant -37000)
@objective(model, Max, 
    45 * x1 + 60 * x2 + 30 * x3 + 18 * x4 - 37000)

# Time constraints (capacity for each department)
@constraint(model, 0.30 * x1 + 0.30 * x2 + 0.25 * x3 + 0.15 * x4 <= 1000)  # Cutting
@constraint(model, 0.25 * x1 + 0.35 * x2 + 0.30 * x3 + 0.10 * x4 <= 1000)  # Insulation
@constraint(model, 0.45 * x1 + 0.50 * x2 + 0.40 * x3 + 0.22 * x4 <= 1000)  # Sewing
@constraint(model, 0.15 * x1 + 0.15 * x2 + 0.20 * x3 + 0.10 * x4 <= 1000)  # Packaging

# Solve the model
optimize!(model)
```

```julia
# Check if the solution is optimal
if termination_status(model) == MOI.OPTIMAL
    println("Optimal solution found:")
    println("Chamarras (x1): ", value(x1))
    println("Relleno de plumas (x2): ", value(x2))
    println("Pantalones (x3): ", value(x3))
    println("Guantes (x4): ", value(x4))
    println("Maximized profit: ", objective_value(model))
else
    println("No optimal solution found.")
end
```

```julia
# Importing JuMP and GLPK
using JuMP
using GLPK

# Create the model with the GLPK solver
model = Model(GLPK.Optimizer)

# Decision variables with integer constraints
@variable(model, x1 >= 0, Int)  # chamarras, integer
@variable(model, x2 >= 0, Int)  # relleno de plumas, integer
@variable(model, x3 >= 0, Int)  # pantalones, integer
@variable(model, x4 >= 0, Int)  # guantes, integer

# Objective function: maximize profit (adjusted coefficients and constant -37000)
@objective(model, Max, 
    45 * x1 + 60 * x2 + 30 * x3 + 18 * x4 - 37000)

# Time constraints (capacity for each department)
@constraint(model, 0.30 * x1 + 0.30 * x2 + 0.25 * x3 + 0.15 * x4 <= 1000)  # Cutting
@constraint(model, 0.25 * x1 + 0.35 * x2 + 0.30 * x3 + 0.10 * x4 <= 1000)  # Insulation
@constraint(model, 0.45 * x1 + 0.50 * x2 + 0.40 * x3 + 0.22 * x4 <= 1000)  # Sewing
@constraint(model, 0.15 * x1 + 0.15 * x2 + 0.20 * x3 + 0.10 * x4 <= 1000)  # Packaging

# Solve the model
optimize!(model)

# Check if the solution is optimal
if termination_status(model) == MOI.OPTIMAL
    println("Optimal integer solution found:")
    println("Chamarras (x1): ", value(x1))
    println("Relleno de plumas (x2): ", value(x2))
    println("Pantalones (x3): ", value(x3))
    println("Guantes (x4): ", value(x4))
    println("Maximized profit: ", objective_value(model))
else
    println("No optimal solution found.")
end
```

En este caso, sí se obtiene la misma solución óptima. En general, imponer restricciones de números enteros en un problema de programación lineal puede llevar a soluciones diferentes a las de la versión no entera (continua).  En este problema, la solución continua produce naturalmente valores enteros para todas las variables (o muy cercanos a los enteros), entonces la solución entera es la misma que la continua.

### Ejercicio #2

```julia
using Pkg
Pkg.add("GLPK")
Pkg.add("DataFrames")
Pkg.add("Plots")
```

```julia
# Importar las librerías necesarias
using JuMP, GLPK, DataFrames, Plots

# Parámetros del problema
demanda = [180, 250, 190, 140, 220, 250]  # Demanda mensual
costo_produccion = [50, 45, 55, 52, 48, 50]  # Costo de producción por ventana en cada mes
costo_inventario = [8, 10, 10, 10, 8, 8]  # Costo de almacenamiento por ventana
capacidad = 225  # Capacidad máxima de producción mensual
```

```julia
# Crear el modelo de programación lineal
modelo = Model(GLPK.Optimizer)

# Variables de decisión
@variable(modelo, 0 <= produccion[i=1:6] <= capacidad)  # Producción en cada mes
@variable(modelo, inventario[1:6] >= 0)  # Inventarios al final de cada mes
```

```julia
# Función objetivo: Minimizar los costos de producción y almacenamiento
@objective(modelo, Min, sum(costo_produccion[i] * produccion[i] for i in 1:6) + 
                       sum(costo_inventario[i] * inventario[i] for i in 1:6))

# Restricciones de inventario
@constraint(modelo, inventario[1] == produccion[1] - demanda[1])  # Inventario inicial
for i in 2:6
    @constraint(modelo, inventario[i] == inventario[i-1] + produccion[i] - demanda[i])
end
```

```julia
# Resolver el problema
optimize!(modelo)

# Imprimir resultados
println("Producción óptima por mes:")
produccion_optima = [value(produccion[i]) for i in 1:6]
println(produccion_optima)

println("\nInventario final por mes:")
inventario_optimo = [value(inventario[i]) for i in 1:6]
println(inventario_optimo)

println("\nCosto total:")
costo_total_optimo = objective_value(modelo)
println("\$$(costo_total_optimo)")
```

```julia
# Comparación con solución no óptima
produccion_no_optima = demanda
inventario_no_optimo = [0 for _ in 1:6]

costo_total_no_optimo = sum(costo_produccion[i] * produccion_no_optima[i] for i in 1:6)

# Mostrar los costos
df = DataFrame(Mes=1:6,
    Produccion_Optima=produccion_optima,
    Inventario_Optimo=inventario_optimo,
    Produccion_No_Optima=demanda,
    Costo_Produccion_Optima=[costo_produccion[i] * produccion_optima[i] for i in 1:6],
    Costo_Inventario_Optima=[costo_inventario[i] * inventario_optimo[i] for i in 1:6],
    Costo_Produccion_No_Optima=[costo_produccion[i] * demanda[i] for i in 1:6],
    Costo_Inventario_No_Optima=[0 for _ in 1:6]
)

println("\nTabla de costos comparativos:")
show(df, allrows=true)
```

```julia
# Diagrama de producción y inventario óptimo
plot(1:6, produccion_optima, label="Producción Óptima", lw=2, marker=:circle, xlabel="Mes", ylabel="Ventanas", title="Producción Óptima y Inventario")
plot!(1:6, inventario_optimo, label="Inventario Óptimo", lw=2, marker=:square)

# Guardar gráfico
savefig("produccion_inventario_optimo.png")
```

```julia
# Crear un nuevo modelo para variables enteras
modelo_entero = Model(GLPK.Optimizer)

# Variables de decisión enteras
@variable(modelo_entero, 0 <= produccion[i=1:6] <= capacidad, Int)  # Producción entera en cada mes
@variable(modelo_entero, inventario[1:6] >= 0, Int)  # Inventarios enteros al final de cada mes

# Función objetivo: Minimizar los costos totales
@objective(modelo_entero, Min, sum(costo_produccion[i] * produccion[i] for i in 1:6) + 
                             sum(costo_inventario[i] * inventario[i] for i in 1:6))

# Restricciones de inventario
@constraint(modelo_entero, inventario[1] == produccion[1] - demanda[1])
for i in 2:6
    @constraint(modelo_entero, inventario[i] == inventario[i-1] + produccion[i] - demanda[i])
end

# Resolver el problema
optimize!(modelo_entero)

# Imprimir resultados para el modelo entero
println("Producción óptima con variables enteras:")
produccion_optima_entera = [value(produccion[i]) for i in 1:6]
println(produccion_optima_entera)

println("\nInventario final con variables enteras:")
inventario_optimo_entero = [value(inventario[i]) for i in 1:6]
println(inventario_optimo_entero)

println("\nCosto total con variables enteras:")
costo_total_entero = objective_value(modelo_entero)
println("\$$(costo_total_entero)")
```

```julia
# Comparación entre continuo y entero
println("\nComparación entre costo continuo y entero:")
println("Costo total (continuo): \$$(costo_total_optimo)")
println("Costo total (entero): \$$(costo_total_entero)")
```

```julia
using JuMP
using HiGHS
```

#### a) Formular el problema de programación lineal

```julia
model = Model(HiGHS.Optimizer)

# Definir las variables de decisión (número de unidades de cada tipo, continuas inicialmente)
@variable(model, sencillas >= 0)
@variable(model, dobles >= 0)
@variable(model, triples >= 0)
@variable(model, cuadruples >= 0)

# Definir la función objetivo (maximizar la recaudación de impuestos)
@objective(model, Max, 
    1000 * sencillas + 
    1900 * dobles + 
    2700 * triples + 
    3400 * cuadruples
)

# Restricción de área disponible
@constraint(model, 
    0.18 * sencillas + 
    0.28 * dobles + 
    0.4 * triples + 
    0.5 * cuadruples <= 63.75 * 0.85
)

# Restricción de financiación máxima
@constraint(model, 
    50000 * sencillas + 
    70000 * dobles + 
    130000 * triples + 
    160000 * cuadruples <= 15000000
)

# Restricciones de proporciones mínimas
@constraint(model, triples + cuadruples >= 0.25 * (sencillas + dobles + triples + cuadruples))
@constraint(model, sencillas >= 0.20 * (sencillas + dobles + triples + cuadruples))
@constraint(model, dobles >= 0.10 * (sencillas + dobles + triples + cuadruples))
```

#### b) Resolver el problema usando la librería JuMP, en variables continuas, y determinar la distribución óptima

```julia
# Optimizar el modelo en variables continuas usando HiGHS
set_optimizer(model, HiGHS.Optimizer)
optimize!(model)

# Mostrar los resultados continuos
println("Resultados en variables continuas:")
println("Unidades sencillas: ", value(sencillas))
println("Unidades dobles: ", value(dobles))
println("Unidades triples: ", value(triples))
println("Unidades cuádruples: ", value(cuadruples))

# Mostrar la recaudación máxima de impuestos con variables continuas
println("Recaudación de impuestos máxima (continuas): ", objective_value(model))
```

#### c) Resolver el problema en variables enteras, y comparar la solución con la de (b)
